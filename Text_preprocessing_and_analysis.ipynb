# Tokenization

```python
text = "The Indian Premier League (IPL) (also known as the TATA IPL for sponsorship reasons) is a men's Twenty20 (T20) cricket league held annually in India. It is contested by ten teams based in seven cities and three states of India.[1][2] The league was founded by the Board of Control for Cricket in India (BCCI) in 2007. Brijesh Patel is the incumbent chairman of the IPL.[3] The competition is usually held annually in summer (between March and May) and has an exclusive window in the ICC Future Tours Programme, meaning that international cricket does not take place during the IPL season.[4]"
print(text)
```

```python
print(text.split())
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(text)
print(tokens)
```

# Stopwords

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopword_list = stopwords.words("english")
print(stopword_list)
```

```python
clean_tokens = [token for token in tokens if token not in stopword_list]
print(clean_tokens)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
clean_tokens = [lemmatizer.lemmatize(token) for token in clean_tokens]
print(clean_tokens)
```

```python
def preprocess_text(text):
  # convert text to lower case and perform tokenization
  tokens = word_tokenize(text.lower())
  # stopword removal
  clean_tokens = [token for token in tokens if token not in stopword_list]
  # remove punctuation
  clean_tokens = [token for token in clean_tokens if token not in punctuation]
  # lemmatization
  lemmatizer = WordNetLemmatizer()
  clean_tokens = [lemmatizer.lemmatize(token) for token in clean_tokens]
  return clean_tokens
```

```python
clean_tokens = preprocess_text(text)
print(clean_tokens)
```

```python
sachin_text = "Sachin, born 24 April 1973 is an Indian former international cricketer who captained the Indian national team. He is regarded as one of the greatest batsmen in the history of cricket.[4] He is the all-time highest run-scorer in both ODI and Test cricket with more than 18,000 runs and 15,000 runs, respectively.[5] He also holds the record for receiving the most man-of-the-match awards in international cricket.[6]"
print(sachin_text)
```

```python
sachin_clean_tokens = preprocess_text(sachin_text)
print(sachin_clean_tokens)
```

# POS Tagging

```python
nltk.download('averaged_perceptron_tagger')
```

```python
pos_tag = nltk.pos_tag(sachin_clean_tokens)
```

```python
print(pos_tag)
```

```python
for tag in pos_tag:
  if tag[1] == "NN":
    print(tag[0])
```

```python
for tag in pos_tag:
  if tag[1] in ["VB", "VBG", "VBN", "VBP", "VBD", "VBZ"]:
    print(tag[0])
```

# Named Entity Recognition

```python
# spacy
import spacy
```

```python
# load a pretrained langauge model
model = spacy.load('en_core_web_sm')
```

```python
clean_text = " ".join(clean_tokens)
print(clean_text)
```

```python
entities = model(clean_text)
```

```python
for ent in entities.ents:
  print(ent.text, ent.label_)
```
